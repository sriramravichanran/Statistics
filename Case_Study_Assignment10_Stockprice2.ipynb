### Case Study-Stock Price:-
The term stock price refers to the current price that a share of stock is trading for on the market. The stock price of a company depends upon various factors like Market Sentiments News, Fundamental factors, technical factors, etc. These stocks are dynamic in nature and the stock value keeps on changing with the time. Here we have the day closing price for the last 1 year of three stocks i.e Adani Greens, Idea Vodafone and Maruti Suzuki. You as a Manager of Investor firm wants to invest a big amount in one of these shares. Apply the various measures of Descriptive Statistics and analyze these stocks.

Here are some questions you need to answer:-

Generate the Descriptive Statistics for all the three stocks manually and verify the results using Data Analysis tool. Draw histogram of all the three stocks.

Now based upon the results of Descriptive Statistics and histogram, answer the following:

1.How are the returns of these 3 stocks skewed?

2.Arrange the stock's average returns in ascending order.

3.Which out of these stocks is the riskiest?

4.Which stock has given maximum and minimum return on a day?

5.Which stock has given positive returns on more than 50 % of the time?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Case_study_StockPrice_1.csv') 
```

```python
# ==========================================================
# 1Ô∏è DESCRIPTIVE STATISTICS (MANUAL + PANDAS)
# ==========================================================

# Columns for stock prices
price_cols = ["Adani Green", "Idea Vodafone", "Maruti Suzuki"]

# Columns for returns
return_cols = ["Adani Green Return %", "Idea Vodafone Return %", "Maruti Suzuki Return %"]

# --- Manual Descriptive Statistics ---
manual_stats = {}

for col in price_cols:
    data = df[col]
    manual_stats[col] = {
        "Mean": data.mean(),
        "Median": data.median(),
        "Std Dev": data.std(),
        "Variance": data.var(),
        "Min": data.min(),
        "Max": data.max(),
        "Range": data.max() - data.min(),
        "Skewness": data.skew(),
        "Kurtosis": data.kurt()
    }

#puts it into a table format
manual_stats_df = pd.DataFrame(manual_stats)
print("\n MANUAL DESCRIPTIVE STATISTICS:\n")
print(manual_stats_df)
```

```python
# --- Verify using Pandas describe() ---

print("\n PANDAS DESCRIBE() OUTPUT:\n")
print(df[price_cols].describe().T)
```

```python
# 2Ô∏è HISTOGRAMS FOR ALL THREE STOCKS

plt.figure(figsize=(12, 8))

# Adani Green
plt.subplot(3, 1, 1)
plt.hist(df["Adani Green"], bins=20)
plt.title("Adani Green - Stock Price Distribution")

# Idea Vodafone
plt.subplot(3, 1, 2)
plt.hist(df["Idea Vodafone"], bins=20)
plt.title("Idea Vodafone - Stock Price Distribution")

# Maruti Suzuki
plt.subplot(3, 1, 3)
plt.hist(df["Maruti Suzuki"], bins=20)
plt.title("Maruti Suzuki - Stock Price Distribution")

plt.tight_layout() # to avoid overlapping bw charts
plt.show()
```

#### 1.How are the returns of these 3 stocks skewed?

```python
# Graphical representation 
# --- Adani ---
plt.hist(df["Adani Green Return %"], bins=20)
plt.title(f"Adani Green Returns\nSkew = {df['Adani Green Return %'].skew():.3f}")
plt.xlabel("Return %") 
plt.ylabel("Frequency")
plt.show()
```

```python
# --- Idea ---
plt.hist(df["Idea Vodafone Return %"], bins=20)
plt.title(f"Idea Vodafone Returns\nSkew = {df['Idea Vodafone Return %'].skew():.3f}")
plt.xlabel("Return %") 
plt.ylabel("Frequency")
plt.show()
```

```python
# --- Maruti ---
plt.hist(df["Maruti Suzuki Return %"], bins=20)
plt.title(f"Maruti Suzuki Returns\nSkew = {df['Maruti Suzuki Return %'].skew():.3f}")
plt.xlabel("Return %") 
plt.ylabel("Frequency")
plt.show()
```

```python
# Columns containing return values
return_cols = ["Adani Green Return %", "Idea Vodafone Return %", "Maruti Suzuki Return %"]

# Calculate skewness
for col in return_cols:
    skew_value = df[col].skew()
    if skew_value > 0:
        skew_type = "Right-skewed (positively skewed)"
    elif skew_value < 0:
        skew_type = "Left-skewed (negatively skewed)"
    else:
        skew_type = "Perfectly symmetrical"
    
    print(f"{col}: {skew_value:.3f} ‚Üí {skew_type}")
```

#### 2.Arrange the stock's average returns in ascending order.

```python
# Calculate mean returns
mean_returns = {
    "Adani Green Return": df["Adani Green Return %"].mean(),
    "Idea Vodafone Return": df["Idea Vodafone Return %"].mean(),
    "Maruti Suzuki Return": df["Maruti Suzuki Return %"].mean()
}

# Sort in ascending order
sorted_means = sorted(mean_returns.items(), key=lambda x: x[1])

print("Average Returns (Ascending Order):")
for stock, avg in sorted_means:
    print(f"{stock}: {avg:.4f}")
```

#### 3.Which out of these stocks is the riskiest?

```python
# Calculate standard deviation of returns
# Stores results in a dictionary
risk = {
    "Adani Green Return": df["Adani Green Return %"].std(),  # Computes std deviation of each stock
    "Idea Vodafone Return": df["Idea Vodafone Return %"].std(),
    "Maruti Suzuki Return": df["Maruti Suzuki Return %"].std()
}

# Find riskiest stock
riskiest_stock = max(risk, key=risk.get) #Uses max() to find which has the highest risk
# .get returns the value (standard deviation) for each stock name (key) in the dictionary.

print("Standard Deviations:", risk)
print("\nRiskiest Stock:", riskiest_stock)
```

To determine which stock is the riskiest, we compare the standard deviation of returns.
Higher standard deviation ‚Üí higher volatility ‚Üí higher risk

#### 4.Which stock has given maximum and minimum return on a day?

```python
# Find maximum return across all 3 stock return columns
max_return = df[["Adani Green Return %", 
                 "Idea Vodafone Return %", # idxmax() finds which stock had the highest among them
                 "Maruti Suzuki Return %"]].max().idxmax() #max() finds the highest daily return 'PER' stock.

min_return = df[["Adani Green Return %", 
                 "Idea Vodafone Return %", 
                 "Maruti Suzuki Return %"]].min().idxmin()

print("Stock with Maximum Return on a Day:", max_return)
print("Stock with Minimum Return on a Day:", min_return)
```

#### 5.Which stock has given positive returns on more than 50 % of the time?

```python
returns = {
    "Adani Green Return": df["Adani Green Return %"],
    "Idea Vodafone Return": df["Idea Vodafone Return %"],
    "Maruti Suzuki Return": df["Maruti Suzuki Return %"]
}

for stock, data in returns.items():
    positive_return = (data > 0).mean()   # % of positive days
    print(stock, "‚Üí Positive Days % =", round(positive_return * 100, 2))


    if positive_return > 0.5:
        print("‚úî", stock, "has positive returns more than 50% of the time\n")
    else:
        print("‚úò", stock, "does NOT have positive returns more than 50% of the time\n")
```
